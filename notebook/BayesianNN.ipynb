```python
import sys
sys.path.append('../')
from models.BayesianNN import BayesianNN
import numpy as np
from utils.plot import plot
import matplotlib.pyplot as plt
from utils.gen_data import gen_data
```

```python
NUM_MODELS = 100
NUM_EPOCHS = 200
BATCH_SIZE = 128
data_x, data_y = gen_data()
```

```python
bdl = BayesianNN(num_models=NUM_MODELS, num_epochs=NUM_EPOCHS, batch_size=BATCH_SIZE, task="data_uncertainty")
x = np.linspace(-4, 4, 100)
mean, variance = bdl.fit_pred(data_x, data_y, x[:, np.newaxis])
```

```python
plot(mean, np.abs(variance), data_x, data_y, x, "data uncdrtainty")
```
